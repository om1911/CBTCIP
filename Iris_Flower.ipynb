```python
#now let us split our data into train and test for accuracy measurement

from sklearn.model_selection import train_test_split
X = df.iloc[:,1:5]
print(X)
X_train,X_test,Y_train,Y_test = train_test_split(X,df.Species,test_size=0.2,random_state=0)
```

```python
#now let us move forward for classification algorithm implementation
#there are many classification algorithm available to choose from such as Support vector machine, logistic regression, K-Nearest Neighbors
#Decision tree etc.. but here i prefer SVM(Support Vector Machine) as our data is small and there are 3 categories to classify from

from sklearn.svm import SVC

classifier = SVC()

classifier.fit(X_train, Y_train)
```

```python
Y_pred = classifier.predict(X_test)
com = pd.DataFrame({"Y_pred" : Y_pred, "Y_test" : Y_test.tolist()})
com
```

```python
from sklearn.metrics import accuracy_score

#accuracy = accuracy_score(Y_test, Y_pred)
print("Accuracy : ",accuracy_score(Y_test, Y_pred)*100)
```
