# **Unemployment analysis**

```python
#let us first get our dataset in this Google colab notebook by using files from google.colab library
#through which we can upload our file and use it to retrieve dataset to conver it into dataframe

from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
import io

df = pd.read_csv(io.BytesIO(uploaded['Unemployment_Rate_upto_11_2020 - Unemployment_Rate_upto_11_2020.csv']))
df
```

```python
#now let us check that our dataset contains any missing values
import missingno as ms

ms.matrix(df)
ms.bar(df)
```

```python
df.isnull().sum()
```

```python
#now let us rename our columns for better understanding and readability
df.columns= ["States","Date","Frequency", "Estimated Unemployment Rate", "Estimated Employed", "Estimated Labour Participation Rate",
               "Region","longitude","latitude"]
df.head()
```

```python
#now let's plot heatmap for finding correlation
import seaborn as sns

cr = df.corr()
sns.heatmap(cr,annot=True)
```

```python
sns.histplot(x="Estimated Unemployment Rate", hue="Region", data=df)
```

```python
sns.histplot(x="Estimated Employed", hue="Region", data=df)
```

```python
sns.displot(data=df, x="Estimated Unemployment Rate", hue="Region", kind="kde")
```

```python
sns.displot(data=df, x="Estimated Employed", hue="Region", kind="kde")
```

```python
sns.histplot(x="Region", y="Estimated Unemployment Rate" , hue="Region", data=df)
```

```python
import plotly.express as px

unemploment = df[["States", "Region", "Estimated Unemployment Rate"]]
figure = px.sunburst(unemploment, path=["Region", "States"],
                     values="Estimated Unemployment Rate",
                     width=700, height=700, color_continuous_scale="RdY1Gn")
figure.show()
```
